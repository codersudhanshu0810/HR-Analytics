# Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Dataset

```python
df = pd.read_csv('HR-Employee-Attrition.csv')
df
```

```python
pd.set_option('display.max_columns', None)
df
```

```python
df = df.drop(columns = ['DistanceFromHome', 'EmployeeCount', 'EmployeeNumber', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager'])
df
```

```python
df = df.drop(columns = 'NumCompaniesWorked')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Attrition'] = df['Attrition'].str.replace('Yes', 'True').replace('No', 'False')
df
```

```python
df['Attrition'] = df['Attrition'].astype(bool)
df.info()
```

```python
df = df.drop_duplicates()
df
```

```python
edf = df['Department'].value_counts()
edf = edf.reset_index()
edf
```

# Data Visualization

```python
plt.figure(figsize = (15, 5))
plt.plot(edf['Department'], edf['count'])
plt.title('Count of workers according to Department')
plt.xlabel('Department Types')
plt.ylabel('Count of Department Workers')
plt.savefig('COUNT OF DEPARTMENT WORKERS.png', format='png')
```

```python
df
```

```python
bdf = df['BusinessTravel'].value_counts()
bdf = bdf.reset_index()
bdf
```

# Data Visualization

```python
plt.figure(figsize = (15, 5))
plt.plot(bdf['BusinessTravel'], bdf['count'])
plt.title('Count of workers who Travels')
plt.xlabel('Business Travel Types')
plt.ylabel('Count of Travelling Workers')
plt.savefig('COUNT OF TRAVEL WORKERS.png', format='png')
```

```python
df
```

```python
fdf = df.groupby('JobRole').agg(TotalIncome = ('MonthlyIncome', 'sum'))
fdf = fdf.reset_index().sort_values(by = 'TotalIncome', ascending = False)
fdf
```

# Data Visualization

```python
plt.figure(figsize = (25, 5))
plt.bar(fdf['JobRole'], fdf['TotalIncome'])
plt.title('Total Income according to Job Role')
plt.xlabel('Job Roles Types')
plt.ylabel('Total Income according to Job Role')
plt.savefig('TOTAL INCOME IN JOB ROLE.png', format='png')
```

```python
df
```

```python
mdf = df.groupby('MaritalStatus').agg(StatusEarn = ('MonthlyRate', 'sum'))
mdf = mdf.reset_index().sort_values(by='StatusEarn', ascending = False)
mdf
```

# Data Visualization

```python
plt.figure(figsize = (20, 5))
plt.plot(mdf['MaritalStatus'], mdf['StatusEarn'])
plt.title('Total Income according to Marital Status')
plt.xlabel('Marital Status Types')
plt.ylabel('Total Income according to Marital Status')
plt.savefig('TOTAL INCOME IN MARITAL STATUS.png', format='png')
```

```python
df
```

```python
gdf = df.groupby('Gender').agg(GenderEarn = ('MonthlyRate', 'sum'))
gdf = gdf.reset_index().sort_values(by = 'GenderEarn', ascending = False)
gdf
```

# Data Visualization

```python
plt.figure(figsize=(4, 6))
plt.pie(gdf['GenderEarn'], labels = gdf['Gender'] , autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Gender Pie Chart')
plt.savefig('GENDER EARN PIE CHART.png', format = 'png')
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.savefig('CORRELATION HEATMAP.png', format = 'p')
```
